```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame([{"nb_chambre": 2, "prix": 10}, {"nb_chambre": 4, "prix": 20}])
data
```

```python
plt.scatter(data["nb_chambre"], data["prix"], c="red")
plt.plot(data["nb_chambre"], data["prix"])
plt.xlabel("Nombre de chambre")
plt.ylabel("Prix")
plt.title("Prix en fonction du nombre de chambre")
plt.show()
```
